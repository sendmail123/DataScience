```
img = cv2.imread("dog_backpack.jpg")
```

```
plt.imshow(img)
```

```
fix_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```
plt.imshow(fix_img)
```

```
fl = cv2.flip(fix_img,-1)
plt.imshow(fl)
```

```
plt.imshow(cv2.rectangle(fix_img,pt1=(220,400),pt2=(550,750),color=(255,0,0), thickness=10))
```

```
vertices = np.array([[300,400],[200,700],[600,700]],dtype = np.int32)
```

```
vertices.shape
```

```
pts=vertices.reshape((-1,1,2))
```

```
pts.shape
```

```
cv2.polylines(fix_img,[pts], isClosed=True,color= (255,0,0), thickness = 10)
plt.imshow(fix_img)
```

```
cv2.fillPoly(fix_img,[pts], color= (255,0,0))
plt.imshow(fix_img)
```
