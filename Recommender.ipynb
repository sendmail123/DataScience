```
Movielens=pd.DataFrame(index=userList,
            columns=itemList)
```

```
Movielens.iloc[0:4,:3]
```

```
a=0
for i in userList:
    for j in df[df['user']==i]['item']:
        Movielens.iloc[a][j]=df[(df['user']==i)&(df['item']==j)]['rating'].item()
    a=a+1
```
