```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
full = cv2.imread('/Users/vishalsingh/Desktop/Data/ML/ComputerVision/Computer-Vision-with-Python/DATA/sammy.jpg')
full = cv2.cvtColor(full,cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(full)
```

```python
face = cv2.imread('/Users/vishalsingh/Desktop/Data/ML/ComputerVision/Computer-Vision-with-Python/DATA/sammy_face.jpg')
face = cv2.cvtColor(face,cv2.COLOR_BGR2RGB)
                  
```

```python
plt.imshow(face)
```

```python
methods = ['cv2.TM_CCOEFF', 'cv2.TM_CCOEFF_NORMED', 'cv2.TM_CCORR','cv2.TM_CCORR_NORMED', 'cv2.TM_SQDIFF', 'cv2.TM_SQDIFF_NORMED']
```

```python
for m in methods:
    
    full_copy = full.copy()
    
    method = eval(m)
    
    #templete Matching
    
    res = cv2.matchTemplate(full_copy,face,method)
    min_val,max_val,min_loc,max_loc = cv2.minMaxLoc(res)
    
    if method in [cv2.TM_SQDIFF, cv2.TM_SQDIFF_NORMED]:
        top_left = min_loc
    else:
        top_left=max_loc
        
    height,width,channels = face.shape
    
    bottom_right = (top_left[0]+width,top_left[1]+height)
    
    cv2.rectangle(full_copy,top_left,bottom_right,color=(0,0,255),thickness=10)
    
    plt.subplot(121)
    plt.imshow(res)
    plt.title('Heat Map of Templete Matching')
    
    plt.subplot(122)
    plt.imshow(full_copy)
    plt.title('Detection of Templete')
    #Title with the method
    plt.suptitle(m)
    
    plt.show()
    
    print('\n')
    print('\n')
                                                
```
