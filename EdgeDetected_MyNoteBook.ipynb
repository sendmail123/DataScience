```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img = cv2.imread('/Users/vishalsingh/Desktop/Data/ML/ComputerVision/Computer-Vision-with-Python/DATA/sammy_face.jpg')
```

```python
plt.imshow(img)
```

```python
edges = cv2.Canny(image=img, threshold1=50,threshold2=255)
```

```python
plt.imshow(edges)
```

```python
med_val = np.median(img)
```

```python
med_val
```

```python
lower = int(max(0,0.7*med_val))
upper = int(min(255,1.3*med_val))
```

```python
edges = cv2.Canny(image=img, threshold1=lower,threshold2=upper+100)
plt.imshow(edges)
```

```python
blur_img = cv2.blur(img,ksize=(5,5))
```

```python
edges = cv2.Canny(image=blur_img, threshold1=lower,threshold2=upper+100)


plt.subplot(122)
plt.imshow(img)
plt.title('Original Image')

plt.subplot(121)
plt.imshow(edges)
plt.title('Edge')



  
plt.show()
print('\n')
print('\n')
    
```
