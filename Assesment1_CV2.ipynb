```
import cv2
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```
def display_img(img):
    fig = plt.figure(figsize=(12,10))
    ax = fig.add_subplot(111)
    ax.imshow(img)
```

```
img = cv2.imread('giraffes.jpg',0)
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```
display_img(img)
```

```
img.shape
```

```
ret, thrsh = cv2.threshold(img,127,255,cv2.THRESH_BINARY)
```

```
display_img(thrsh)
```

```
img_hs = cv2.imread('giraffes.jpg')
img_hs = cv2.cvtColor(img_hs,cv2.COLOR_BGR2HSV) 
```

```
display_img(img_hs)
```

```
kernel = np.ones((4,4),dtype=np.float32)/10
```

```
kernel
```

```
img = cv2.imread('giraffes.jpg')
img =cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
```

```
filter = cv2.filter2D(img,-1,kernel=kernel)
display_img(filter)
```

```
img = cv2.imread('giraffes.jpg',0)
```

```
sobelx = cv2.Sobel(img,cv2.CV_64F,1,0)
display_img(sobelx)
```

```
plt.imshow(sobelx,cmap='gray')
```

```
img = cv2.imread('giraffes.jpg')
```

```
hist = cv2.calcHist([img],[2],None,[256],[0,256])
```

```
plt.plot(hist)
```

```
img = cv2.imread('giraffes.jpg')

color = ('b','g','r')

for i, col in enumerate(color):
  hist = cv2.calcHist([img],[i],None,[256],[0,256])
  plt.plot(hist,color=col)

plt.title('Histogram for Giraffe')
```
